```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(open("q.html"), features="html")
```

```python
lecture_list=soup.find_all('a')
```

```python
lecture_list[0]
```

```python
lecture_list[0].get("href")
```

```python
lecture_links=[]

for l in lecture_list:
    link="https://www.quantopian.com"+l.get("href")
    print(link)
    lecture_links.append(link)
```

```python
l=lecture_links[1]
l
```

```python
import requests
headers = requests.utils.default_headers()

req = requests.get(l, headers)
soup = BeautifulSoup(req.content, features="html")
soup
```

```python
yt_div=soup.find("div", { "class" : "youtube-player" })
yt_div
```

```python
yt_id=yt_div.get("data-id")
yt_id
```

```python
website="https://www.youtube.com/watch?v="
print(website+yt_id)
```

```python
soup.find('title').string.strip("\n")
```

```python
videos_links=[]
lectures=[]

for l in lecture_links:
    print(l)    
    req = requests.get(l, headers)
    soup = BeautifulSoup(req.content, features="html")
    try:
        yt_div=soup.find("div", { "class" : "youtube-player" })
        yt_id=yt_div.get("data-id")
        yt_link=website+yt_id
        
    except:
        yt_link="Exercise"
        
    videos_links.append(yt_link)
    lecture=soup.find('title').string.strip("\n")
    lectures.append(lecture)
```

```python
df_dict={
    "Lectures":lectures,
    "Lecture Links":lecture_links,
    "Video":videos_links,
}
df=pd.DataFrame(df_dict)
df.to_csv("Quantopian.csv")
df.to_excel("Quantopian.xlsx")
```
